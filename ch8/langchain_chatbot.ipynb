## 모델 선언

```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model="gpt-4o-mini")
```

## 첫 번째 대화

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content="안녕? 나는 김정은이야.")])
```
